```python
# data spliting
```

```python
x=df[["AI_Usage_Hours","Sleep_Hours","Cognitive_Load","Coffee_Intake","Stress_Level"]]
y=df["Task_Success_Rate"]
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

reg_LR = LinearRegression()

reg_LR.fit(x_train,y_train)

y_pred = reg_LR.predict(x_test)

s_LR = r2_score(y_test,y_pred)
print("r2 Score = ",s_LR)

P1_LR = reg_LR.predict([[6,6.2,87,3,92]])
print('prediction 1 = ',P1_LR)

P2_LR = reg_LR.predict([[6,6.6,86,4,100]])
print('prediction 2 = ',P2_LR)

P3_LR = reg_LR.predict([[1,6.1,28,5,36]])
print('prediction 3 = ',P3_LR)
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

reg_DT = DecisionTreeRegressor()

reg_DT.fit(x_train,y_train)

y_pred = reg_DT.predict(x_test)

s_DT = r2_score(y_test,y_pred)
print("r2 Score = ",s_DT)

P1_DT = reg_DT.predict([[6,6.2,87,3,92]])
print('prediction 1 = ',P1_DT)

P2_DT= reg_DT.predict([[6,6.6,86,4,100]])
print('prediction 2 = ',P2_DT)

P3_DT = reg_DT.predict([[1,6.1,28,5,36]])
print('prediction 3 = ',P3_DT)
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

reg_RF = RandomForestRegressor()

reg_RF.fit(x_train,y_train)

y_pred = reg_RF.predict(x_test)

s_RF = r2_score(y_test,y_pred)
print("r2 Score = ",s_RF)

P1_RF= reg_RF.predict([[6,6.2,87,3,92]])
print('prediction 1 = ',P1_RF)

P2_RF = reg_RF.predict([[6,6.6,86,4,100]])
print('prediction 2 = ',P2_LR)

P3_RF = reg_RF.predict([[1,6.1,28,5,36]])
print('prediction 3 = ',P3_RF)
```

```python
#Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

reg_GBR = GradientBoostingRegressor()

reg_GBR.fit(x_train,y_train)

y_pred = reg_GBR.predict(x_test)

s_GBR = r2_score(y_test,y_pred)
print("r2 Score = ",s_GBR)

P1_GBR = reg_GBR.predict([[6,6.2,87,3,92]])
print('prediction 1 = ',P1_GBR)

P2_GBR = reg_GBR.predict([[6,6.6,86,4,100]])
print('prediction 2 = ',P2_GBR)

P3_GBR = reg_GBR.predict([[1,6.1,28,5,36]])
print('prediction 3 = ',P3_GBR)
```

```python
#SVR Regressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score

reg_SVR = SVR()

reg_SVR.fit(x_train,y_train)

y_pred = reg_SVR.predict(x_test)

s_SVR = r2_score(y_test,y_pred)
print("r2 Score = ",s_SVR)

P1_SVR = reg_SVR.predict([[6,6.2,87,3,92]])
print('prediction 1 = ',P1_SVR)

P2_SVR = reg_SVR.predict([[6,6.6,86,4,100]])
print('prediction 2 = ',P2_SVR)

P3_SVR = reg_SVR.predict([[1,6.1,28,5,36]])
print('prediction 3 of 170 = ',P3_SVR)
```

```python
#KNeighbors Regressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

reg_KNN = KNeighborsRegressor()

reg_KNN.fit(x_train,y_train)

y_pred = reg_KNN.predict(x_test)

s_KNN = r2_score(y_test,y_pred)
print("r2 Score = ",s_KNN)

P1_KNN = reg_KNN.predict([[6,6.2,87,3,92]])
print('prediction 1 = ',P1_KNN)

P2_KNN = reg_KNN.predict([[6,6.6,86,4,100]])
print('prediction 2 = ',P2_KNN)

P3_KNN = reg_KNN.predict([[1,6.1,28,5,36]])
print('prediction 3 = ',P3_KNN)
```
